```python
def newvoting(x_test, x_train, y_train):
    model = baseline_model()
    model.fit(x_train, y_train, validation_data=(x_val, y_val), epochs=epochs, batch_size=batch_size, verbose=2)

    # Predict
    y_pred = model.predict(x_test)
    rounded = [numpy.round(x) for x in y_pred]
    # print(numpy.shape(rounded))

    y_test = numpy.zeros((800, 1))
    for i in range(800):
    # import pdb;pdb.set_trace()
        val, = numpy.nonzero(rounded[i][:])
        if bool(val):
            # print(val)
            y_test[i, 0] = val
        else:
            y_test[i, 0] = 0
    return y_test
```

```python
y_test11 = pd.DataFrame(y_test1, columns = ['y1'])
y_test22 = pd.DataFrame(y_test2, columns = ['y2'])
y_test33 = pd.DataFrame(y_test3, columns = ['y3'])
```

```python
df = pd.DataFrame()

df['y1'] = y_test11
df['y2'] = y_test22
df['y3'] = y_test33
```

```python
df = df.astype(int)
```

```python
mode = df.mode(axis=1)
```

```python
mode = mode.values.ravel()
```

```python
def order(frame,var):
    varlist =[w for w in frame.columns if w not in var]
    frame = frame[var+varlist]
    return frame
```

```python
d = {'id': range(len(mode)), 'digit': mode.astype(int)}
df_output = pd.DataFrame(data=d)
df_output = order(df_output,['id','digit'])
df_output.to_csv(
     r'submissions.csv',
     index=False
)
```
